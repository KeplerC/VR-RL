```python
import pandas as pd
import os
```

```python
import os
all_f = [[f for f in os.listdir() if f.startswith("1")][1]]
dataframes = []
for f in all_f:
    dataframes.append(pd.read_csv(f, names=list('0123456789012345678901234567')))
```

```python
df = pd.concat(dataframes, axis = 0)
```

```python
df
```

```python
output = []
for analyzer in pd.unique(df['0']):
    temp_mat = df.loc[df['0'] == analyzer]
    temp_mat['1'] = pd.to_datetime(temp_mat['1'])
    temp_mat.set_index('1').resample(rule='20ms').sum().T
    output.append(temp_mat)
```

```python
def roundTime(dt, roundTo = 20):
    roundTo = roundTo * 1000
    #dt = dt.replace(microsecond = dt.microsecond // roundTo * roundTo)
    dt = dt.microsecond // roundTo
    return dt
import numpy as np
output = []
i = 0
for analyzer in pd.unique(df['0']):
    temp_mat = df.loc[df['0'] == analyzer]
    temp_mat['0'] = np.ones((temp_mat.shape[0], 1), dtype=np.int) * i
    temp_mat['1'] = [roundTime(pd.to_datetime(t)) for t in temp_mat['1']]
    #df.set_index('1').resample(rule='20ms').sum().T
    temp_mat.set_index('1').sum().T
    output.append(temp_mat)
a = pd.concat(output, axis = 0).sort_values(by='1')
```
